# Feature Extraction

```python
import pandas as pd
import numpy as np
import scipy.stats
from scipy import signal
import pywt

def extract_features(dict_of_dfs):
    all_features = []
    
    for key, df in dict_of_dfs.items():
        features = {}

        for column in ['ax', 'ay', 'az', 'gx', 'gy', 'gz']:
            col_data = df[column]

            # Calculate Mean
            features[f'{column}_mean'] = col_data.mean()

            # Calculate Standard Deviation
            features[f'{column}_std'] = col_data.std()

            # Calculate Kurtosis
            features[f'{column}_kurtosis'] = col_data.kurtosis()

            # Calculate Minimum
            features[f'{column}_min'] = col_data.min()

            # Calculate Maximum
            features[f'{column}_max'] = col_data.max()

            # Calculate Skewness
            features[f'{column}_skewness'] = col_data.skew()

            # Calculate 1st quartile
            features[f'{column}_1st_quartile'] = np.percentile(col_data, 25)

            # Calculate Median
            features[f'{column}_median'] = np.median(col_data)

            # Calculate 3rd quartile
            features[f'{column}_3rd_quartile'] = np.percentile(col_data, 75)

            # Calculate Peak to Peak Amplitude
            features[f'{column}_peak_to_peak'] = col_data.max() - col_data.min()

            # Calculate Root Mean Square Amplitude
            features[f'{column}_rms_amplitude'] = np.sqrt(np.mean(col_data ** 2))

            # Calculate Crest Factor
            features[f'{column}_crest_factor'] = col_data.max() / np.sqrt(np.mean(col_data ** 2))

            # Calculate Zero Crossing
            zero_crossings = len(np.where(np.diff(np.signbit(col_data)))[0])
            features[f'{column}_zero_crossing'] = zero_crossings

            # Calculate Mean Crossing
            mean_crossings = len(signal.find_peaks(np.abs(col_data - col_data.mean()))[0])
            features[f'{column}_mean_crossing'] = mean_crossings

            # Calculate Impulse Factor
            features[f'{column}_impulse_factor'] = col_data.max() / np.sum(np.abs(col_data))

            # Calculate Shape Factor
            coeffs = pywt.wavedec(col_data, wavelet='db2', level=3)
            third_coefficient = coeffs[2]
            features[f'{column}_shape_factor'] = np.abs(np.sum(third_coefficient) / len(third_coefficient))

            # Calculate Clearance Factor
            coeffs_clearance = pywt.wavedec(col_data, wavelet='db2', level=6)
            sixth_coefficient = coeffs_clearance[5]
            features[f'{column}_clearance_factor'] = np.abs(np.sum(sixth_coefficient) / len(sixth_coefficient))

            # Calculate Entropy
            features[f'{column}_entropy'] = scipy.stats.entropy(col_data)

            # Calculate Fast Fourier Transform
            fft_values = np.fft.fft(col_data)
            fft_freq = np.fft.fftfreq(len(col_data))
            max_freq_index = np.argmax(np.abs(fft_values))
            features[f'{column}_fft_max_frequency'] = fft_freq[max_freq_index]
            features[f'{column}_fft_max_amplitude'] = np.abs(fft_values[max_freq_index])

            # Calculate Power Spectral Density
            f, psd = signal.welch(col_data, fs=1)
            features[f'{column}_psd_mean'] = np.mean(psd)
            features[f'{column}_psd_median'] = np.median(psd)

        all_features.append(features)

    features_df = pd.DataFrame(all_features)
    return features_df
```

```python
# Creating a list of dictionaries that are holding 10 dataframes each
dfs_list = [df1s, df2s, df3s, df4s, df5s, df6s, df7s, df8s, df9s, df10s, df11s, df12s, df13s, df14s, df15s, df16s]

# Creating an empty pandas Dataframe
features = pd.DataFrame()

for dfs in dfs_list:
    features_df = extract_features(dfs)
    features = pd.concat([features, features_df], ignore_index=True)
```

# Feature Importance

```python
from sklearn.feature_selection import RFECV
from sklearn.svm import SVC  # Import SVC directly

X = data.iloc[:, :data.shape[1]-6]  # Input (Shape -> (160, 77))
y = data.iloc[:, -1:]   # Output (Shape -> (160, 1))
y = np.ravel(y)

svc = SVC(kernel='linear')  # Use SVC directly without referencing svm
svc.fit(X, y)
selector = RFECV(estimator=svc, step=2, cv=5)
selector = selector.fit(X, y)
z = selector.ranking_ 
q = selector.feature_names_in_
# We can now see the ranking of features that are helpful in predicting the target
print("Ranking of features : \n", z)
```

```python
min_features_to_select = 1
plt.figure()
plt.xlabel("Number of selected features")
plt.ylabel("Cross Validation Accuracy")
plt.plot(
    range(min_features_to_select, len(selector.cv_results_['split0_test_score']) + min_features_to_select),
    selector.cv_results_['split0_test_score'],
)
plt.plot(
    range(min_features_to_select, len(selector.cv_results_['split1_test_score']) + min_features_to_select),
    selector.cv_results_['split1_test_score'],
)
plt.plot(
    range(min_features_to_select, len(selector.cv_results_['split2_test_score']) + min_features_to_select),
    selector.cv_results_['split2_test_score'],
)
plt.plot(
    range(min_features_to_select, len(selector.cv_results_['split3_test_score']) + min_features_to_select),
    selector.cv_results_['split3_test_score'],
)
plt.plot(
    range(min_features_to_select, len(selector.cv_results_['split4_test_score']) + min_features_to_select),
    selector.cv_results_['split4_test_score'],
)
plt.show()
```

### Observation :
We can notice that with 25 features we can achieve approximately 95% accuracy.

```python
res = dict(zip(q,z)) 
selected_features = []

for i,j in res.items():
  if j<=1:
    selected_features.append(i)
print(selected_features)
```

```python
data_subset = pd.DataFrame(data, columns = selected_features)
print(data_subset)
```

# Plotting accuracy v/s num_features for a logistic regression model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data_subset, y, test_size=0.25, random_state=26)

# Initialize an empty list to store ROC-AUC scores for each output
roc_auc_scores = []

# Get the indices of selected features
selected_features_indices = [data_subset.columns.get_loc(feature) for feature in selected_features]

# Iterate over increasing number of features
for num_features in range(1, len(selected_features_indices) + 1):
    # Select the top num_features features
    selected_X_train = X_train.iloc[:, selected_features_indices[:num_features]]
    selected_X_test = X_test.iloc[:, selected_features_indices[:num_features]]

    # Train a Random Forest classifier
    rf_classifier = RandomForestClassifier(n_estimators=20, random_state=42)
    rf_classifier.fit(selected_X_train, y_train)

    # Predict probabilities on the test set
    y_prob = rf_classifier.predict_proba(selected_X_test)

    # Compute ROC-AUC score
    roc_auc = roc_auc_score(y_test, y_prob, multi_class='ovr')  # Use 'ovr' for multiclass classification
    roc_auc_scores.append(roc_auc)

# Plot ROC-AUC score with respect to the number of features used
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(selected_features_indices) + 1), roc_auc_scores, marker='o', linestyle='-')
plt.xlabel('Number of Features')
plt.ylabel('ROC-AUC Score')
plt.title('ROC-AUC Score vs Number of Features')
plt.grid(True)
plt.show()
```

### Observation : 
We note that for 9 features, we are getting ROC_AUC score of around 99% hence we do not really need 25 features to classify these faults.
